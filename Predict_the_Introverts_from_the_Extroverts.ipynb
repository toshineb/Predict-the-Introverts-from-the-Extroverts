```python
label = 'Personality'
train[label].value_counts()
```

```python
from autogluon.tabular import TabularDataset, TabularPredictor
```

```python
predictor = TabularPredictor(label=label,eval_metric ='accuracy',
                            problem_type="binary").fit_pseudolabel(pseudo_data=test,
                                                                   train_data=train,
                                                                   presets='medium_quality',
                                                                    time_limit=3600*9,verbosity=3
                                                      )
results = predictor.fit_summary()
```

```python
predictor.leaderboard()
```

```python
df = predictor.predict(test).to_frame(name=label)
df.head()
```

```python
sample=pd.read_csv('sample_submission.csv')
sample.head()
```

```python
sample[label]=df[label]
sample.head()
```

```python
sample.to_csv('Autogluon_medium_quality_gpu.csv',index=False)
```
